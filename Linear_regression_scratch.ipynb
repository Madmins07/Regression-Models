## ***LINEAR REGRESSION***

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
import tensorflow_datasets as tfds
import random
```

```python
url = 'https://raw.githubusercontent.com/rodmsmendes/reinforcementlearning4fun/master/data/house_prices.csv'
df = pd.read_csv(url)
area=np.array(df['LotArea'])
price=np.array(df['SalePrice'])
plt.plot(area,price,'+')
plt.title("Plot of sales price vs area")
plt.xlabel("Area of house")
plt.ylabel("Price of the house")
plt.show()
```

## Scaling the values down

```python
x=np.copy(area/1000)
y=np.copy(price/1000)
plt.plot(x,y,'r+')
```

```python
def initialise_param():
  w1=np.ones(1460)
  b1=1
  return w1,b1

def update_param(w1,b1,y_pred,x,y,alpha):
  dw = ((y_pred - y)*x).mean()
  db = ((y_pred - y)).mean()

  w1=w1-alpha*dw
  b1=b1-alpha*db
  return w1,b1

def gradient_descent(iter,alpha,x,y):
  w1,b1=initialise_param()
  for i in range(iter):
    y_pred=w1*x+b1
    w1,b1=update_param(w1,b1,y_pred,x,y,alpha)
  return w1,b1
def print_line(w1,b1,x):
  y_plot=w1*x+b1
  plt.plot(x,y_plot)
  plt.plot(x,y,'r+')
  plt.figure(figsize=(5, 5))
  plt.show()

w1,b1=gradient_descent(100,0.001,x,y)
print_line(w1,b1,x)
print(w1*8.450+b1)
```
